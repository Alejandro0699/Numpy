# Extracci贸n

```python
import pandas as pd
import numpy as np
```

```python
#importamos el json
url='https://raw.githubusercontent.com/alura-cursos/challenge2-data-science-LATAM/refs/heads/main/TelecomX_Data.json'
df_telecom=pd.read_json(url)
df_telecom.head()
```

# Transformaci贸n

###Conoce el conjunto de datos

```python
df_telecom.info()
```

```python
#normalizamos el df
df_customer = pd.json_normalize(df_telecom['customer'])
df_phone = pd.json_normalize(df_telecom['phone'])
df_internet= pd.json_normalize(df_telecom['internet'])
df_account = pd.json_normalize(df_telecom['account'])
#eliminamos las columnas customer, phone, internet, account
df_telecom.drop(['customer', 'phone', 'internet', 'account'], axis=1, inplace=True)
#unico o concatenamos las columnas normalizadas
df_telecom_norm = pd.concat([df_telecom, df_customer, df_phone, df_internet, df_account], axis=1)
df_telecom_norm.head()
```

```python
columnas=list(df_telecom_norm.columns)
columnas
```

###Comprobaci贸n de incoherencias en los datos
En este paso, verifica si hay problemas en los datos que puedan afectar el an谩lisis. Presta atenci贸n a valores ausentes, duplicados, errores de formato e inconsistencias en las categor铆as. Este proceso es esencial para asegurarte de que los datos est茅n listos para las siguientes etapas.

```python
df_telecom_norm.info()
```

```python
#detectamos si hay valores nulos
df_telecom_norm.isnull().sum()
```

```python
#detectamos si hay str vacios, me fije en todas las filas una por una y solamente deje las que estan vacios o tienen un espacio en blanco
(df_telecom_norm['Churn']=='').sum()
```

###Manejo de inconsistencias
Ahora que has identificado las inconsistencias, es momento de aplicar las correcciones necesarias. Ajusta los datos para asegurarte de que est茅n completos y coherentes, prepar谩ndolos para las siguientes etapas del an谩lisis.

```python
#ahora trataremos los valores str vacios
df_telecom_norm['Churn']= df_telecom_norm['Churn'].str.strip().replace('',pd.NA)
df_telecom_norm['Charges.Total']= df_telecom_norm['Charges.Total'].str.strip().replace('',pd.NA)
df_telecom_norm.isnull().sum()
```

```python
#eliminaremos las filas que tengan los valores nulos
df_telecom_norm.dropna(subset=['Churn','Charges.Total'], inplace=True)
df_telecom_norm.isnull().sum()
```

```python
df_telecom_norm.info()
```

```python
#creamos una lista con las columnas, que queremos transformar a booleanos
df_bool=['Partner','Dependents','PhoneService','PaperlessBilling','Churn']
df_telecom_norm[df_bool]=df_telecom_norm[df_bool].replace({'Yes': True, 'No':False})
df_telecom_norm[df_bool]=df_telecom_norm[df_bool].astype('bool')
df_telecom_norm.info()
```

```python
#creamos una lista para transformar str
df_str=['gender','InternetService','Contract','PaymentMethod']
df_telecom_norm[df_str]=df_telecom_norm[df_str].astype('string')
#pasamos a float la columna Changes.Total
df_telecom_norm['Charges.Total']=df_telecom_norm['Charges.Total'].astype('float64')
#creamos una lista para cambiar a categoria
df_categoria= ['MultipleLines', 'OnlineSecurity', 'OnlineBackup', 'DeviceProtection', 'TechSupport', 'StreamingTV', 'StreamingMovies']
df_telecom_norm[df_categoria]=df_telecom_norm[df_categoria].astype('category')
df_telecom_norm.info()
```

```python
#Verificamos duplicados
df_telecom_norm.duplicated().sum()
```

###Columna de cuentas diarias
Ahora que los datos est谩n limpios, es momento de crear la columna "Cuentas_Diarias". Utiliza la facturaci贸n mensual para calcular el valor diario, proporcionando una visi贸n m谩s detallada del comportamiento de los clientes a lo largo del tiempo.

 Esta columna te ayudar谩 a profundizar en el an谩lisis y a obtener informaci贸n valiosa para las siguientes etapas.

```python
df_telecom_norm['Cuentas_Diarias']=df_telecom_norm['Charges.Monthly']/30
df_telecom_norm['Cuentas_Diarias']=df_telecom_norm['Cuentas_Diarias'].astype('float64')
df_telecom_norm.head()
```

###Estandarizaci贸n y transformaci贸n de datos (opcional)

Adem谩s, traducir o renombrar columnas y datos hace que la informaci贸n sea m谩s accesible y f谩cil de entender, especialmente cuando se trabaja con fuentes externas o t茅rminos t茅cnicos.
```python
#Procedemos a transformar los valores booleano, que ya tenemos en una variable para pasarlos a numeros binarios
df_telecom_norm[df_bool]=df_telecom_norm[df_bool].astype('int64')
df_telecom_norm.head()
```
